```python
import numpy as np

# Matriz A
A = np.array([[4, 2],
              [1, 3]])

# CÃ¡lculo dos autovalores e autovetores
autovalores, autovetores = np.linalg.eig(A)

autovalores, autovetores
```

IMG_9791.jpeg

```python
for i in range(len(autovalores)):
    Av = np.dot(A, autovetores[:, i])           # Calcula A * v_i
    lambdav = autovalores[i] * autovetores[:, i] # Calcula Î»_i * v_i

    print(f"Autovalor {i+1}: {autovalores[i]}")
    print("A * v:", Av)
    print("Î» * v:", lambdav)
    print()
```

IMG_9792.jpeg

```python
traco_A = np.trace(A)                # Calcula o traÃ§o da matriz A (soma dos elementos da diagonal)
soma_autovalores = np.sum(autovalores)  # Soma dos autovalores calculados anteriormente

traco_A, soma_autovalores
```

IMG_9793.jpeg

```python
det_A = np.linalg.det(A)          # Calcula o determinante da matriz A
produto_autovalores = np.prod(autovalores)  # Calcula o produto dos autovalores

det_A, produto_autovalores
```
